## Import Relevant Library


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load Data

```python
df = pd.read_csv("apple_quality.csv")
df
```

make a copy of the data

```python
data=df.copy()
data
```

```python
data.describe()
```

## checking and removing missing values

```python
data.isnull().sum()
```

```python
data1 = data.dropna(axis = 0)
data1.isnull().sum()
```

## Dropping unwanted rows and columns

```python
data1 = data1.drop(['A_id'], axis = 1)
data1
```

## Creating dummy variables

```python
data1['Quality'] = data['Quality'].map({'good':1, 'bad':0})
data1
```

## Saving the cleaned data

```python
data1.to_csv('AppleStoreClearned.csv', index = False, header = False)
```
